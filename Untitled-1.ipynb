```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Constantes do pêndulo duplo
G = 9.8
L1 = 5.0
L2 = 1.0
M1 = 2.0
M2 = 1.0

def equacoes_pendulo(y, t, L1, L2, M1, M2, G):
    """Define as equações diferenciais para o pêndulo duplo"""
    theta1, w1, theta2, w2 = y

    cos_diff = np.cos(theta1 - theta2)
    sin_diff = np.sin(theta1 - theta2)

    denom = L1 * (M1 + M2 - M2 * cos_diff**2)

    w1_ponto = (M2 * G * np.sin(theta2) * cos_diff - M2 * sin_diff * (L1 * w1**2 * cos_diff + L2 * w2**2) - \
                (M1 + M2) * G * np.sin(theta1)) / denom

    w2_ponto = ((M1 + M2) * (L1 * w1**2 * sin_diff - G * np.sin(theta2) + G * np.sin(theta1) * cos_diff) + \
                M2 * L2 * w2**2 * sin_diff * cos_diff) / (L2 * (M1 + M2 - M2 * cos_diff**2))

    return [w1, w1_ponto, w2, w2_ponto]

def edo_pendulo(estado_inicial, intervalo_tempo):
    """Resolve as equações diferenciais do pêndulo usando odeint."""
    print("Iniciando simulação com odeint...")
    solucao = odeint(equacoes_pendulo, estado_inicial, intervalo_tempo, args=(L1, L2, M1, M2, G))
    print("Simulação concluída.")
    return solucao

def salvar_animacao(ani, filename):
    """Função genérica para salvar a animação."""
    print(f"Salvando animação em {filename}...")
    ani.save(filename, writer='pillow', fps=20) # Aumentando o FPS para 20
    print("Animação salva com sucesso.")

def animar_pendulo(solucao, intervalo_tempo, filename="animacao_trajetoria.gif"):
    """Cria e salva a animação do pêndulo e sua trajetória."""
    theta1 = solucao[:, 0]
    theta2 = solucao[:, 2]

    x1 = L1 * np.sin(theta1)
    y1 = -L1 * np.cos(theta1)
    x2 = x1 + L2 * np.sin(theta2)
    y2 = y1 - L2 * np.cos(theta2)

    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_xlim(-(L1 + L2) * 1.1, (L1 + L2) * 1.1)
    ax.set_ylim(-(L1 + L2) * 1.1, (L1 + L2) * 1.1)
    ax.set_aspect("equal")
    ax.grid()

    line, = ax.plot([], [], "o-", lw=2)
    trace1, = ax.plot([], [], "-", color='gray', lw=1, alpha=0.5)
    trace2, = ax.plot([], [], "-", color='gray', lw=1, alpha=0.5)
    time_text = ax.text(0.05, 0.9, f'Tempo: {intervalo_tempo[0]:.2f} s', transform=ax.transAxes)

    def update(frame):
        thisx = [0, x1[frame], x2[frame]]
        thisy = [0, y1[frame], y2[frame]]

        line.set_data(thisx, thisy)
        trace1.set_data(x1[:frame+1], y1[:frame+1])
        trace2.set_data(x2[:frame+1], y2[:frame+1])
        time_text.set_text(f'Tempo: {intervalo_tempo[frame]:.2f} s')
        return line, trace1, trace2, time_text

    ani = FuncAnimation(fig, update, frames=len(intervalo_tempo), interval=20, blit=True)
    salvar_animacao(ani, filename)

def animar_espaco_fase(solucao, intervalo_tempo, filename="espaco_fase.gif"):
    """Cria e salva a animação do espaço de fase."""
    theta1, w1, theta2, w2 = solucao[:, 0], solucao[:, 1], solucao[:, 2], solucao[:, 3]

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Massa 1
    axes[0].set_xlabel("$\\theta_1$ (rad)")
    axes[0].set_ylabel("$\\omega_1$ (rad/s)")
    axes[0].set_title("Espaço de Fase da Massa 1")
    axes[0].set_xlim(np.min(theta1) - 0.5, np.max(theta1) + 0.5)
    axes[0].set_ylim(np.min(w1) - 0.5, np.max(w1) + 0.5)
    axes[0].grid()

    # Massa 2
    axes[1].set_xlabel("$\\theta_2$ (rad)")
    axes[1].set_ylabel("$\\omega_2$ (rad/s)")
    axes[1].set_title("Espaço de Fase da Massa 2")
    axes[1].set_xlim(np.min(theta2) - 0.5, np.max(theta2) + 0.5)
    axes[1].set_ylim(np.min(w2) - 0.5, np.max(w2) + 0.5)
    axes[1].grid()

    line1, = axes[0].plot([], [], '-', lw=1)
    line2, = axes[1].plot([], [], '-', lw=1)
    time_text = fig.suptitle(f'Tempo: {intervalo_tempo[0]:.2f} s')

    def update(frame):
        line1.set_data(theta1[:frame+1], w1[:frame+1])
        line2.set_data(theta2[:frame+1], w2[:frame+1])
        time_text.set_text(f'Tempo: {intervalo_tempo[frame]:.2f} s')
        return line1, line2, time_text

    ani = FuncAnimation(fig, update, frames=len(intervalo_tempo), interval=20, blit=False)
    salvar_animacao(ani, filename)

if __name__ == '__main__':
    estado_inicial = [np.pi / 2, 0, np.pi / 2, 0]
    dt = 0.02
    intervalo_tempo = np.arange(0, 60, dt)

    solucao = edo_pendulo(estado_inicial, intervalo_tempo)

    # Verifica se a solução é válida antes de tentar animar
    if np.any(np.isnan(solucao)) or np.any(np.isinf(solucao)):
        print("AVISO: A simulação produziu valores inválidos (NaN ou Inf). A animação pode falhar.")
    else:
        animar_pendulo(solucao, intervalo_tempo)
        animar_espaco_fase(solucao, intervalo_tempo)

    plt.show()
```
